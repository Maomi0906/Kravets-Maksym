```python
from ColabTurtle.Turtle import *

initializeTurtle()
speed(13)

# Střed plátna
cx = 400
cy = 225

# Počet paprsků
num_rays = 16
# Počet vrstev (kruhových obručí)
num_rings = 8
# Délka nejdelšího paprsku
max_length = 150

# --- Kreslíme paprsky ---
for i in range(num_rays):
    penup()
    goto(cx, cy)
    pendown()
    setheading(i * (360 / num_rays))
    forward(max_length)

# --- Kreslíme obruče pavučiny ---
# Pavučina bude tvořena "vlnkami" připojenými mezi paprsky
for ring in range(1, num_rings + 1):
    radius = (max_length / num_rings) * ring

    # projdeme paprsky a propojíme je malým obloučkem
    for i in range(num_rays):
        angle = i * (360 / num_rays)
        next_angle = (i + 1) * (360 / num_rays)

        # začátek oblouku
        x1 = cx + radius * 0.95 * math.cos(math.radians(angle))
        y1 = cy + radius * 0.95 * math.sin(math.radians(angle))

        # konec oblouku
        x2 = cx + radius * 0.95 * math.cos(math.radians(next_angle))
        y2 = cy + radius * 0.95 * math.sin(math.radians(next_angle))

        penup()
        goto(x1, y1)
        pendown()

        # místo rovné čáry uděláme lehký oblouček
        mid_x = (x1 + x2) / 2
        mid_y = (y1 + y2) / 2 + 3   # malé prohnutí pro "pavoučí tvar"

        goto(mid_x, mid_y)
        goto(x2, y2)
```

```python
from ColabTurtle.Turtle import *
import math

initializeTurtle()
speed(13)
cx = 400
cy = 225
num_rays = 16
num_rings = 8
max_length = 150

for i in range(num_rays):
    penup()
    goto(cx, cy)
    pendown()
    setheading(i * (360 / num_rays))
    forward(max_length)

for ring in range(1, num_rings + 1):
    radius = (max_length / num_rings) * ring

    for i in range(num_rays):
        angle = i * (360 / num_rays)
        next_angle = (i + 1) * (360 / num_rays)

        x1 = cx + radius * 0.95 * math.cos(math.radians(angle))
        y1 = cy + radius * 0.95 * math.sin(math.radians(angle))

        x2 = cx + radius * 0.95 * math.cos(math.radians(next_angle))
        y2 = cy + radius * 0.95 * math.sin(math.radians(next_angle))

        mid_x = (x1 + x2) / 2
        mid_y = (y1 + y2) / 2 + 3

        penup()
        goto(x1, y1)
        pendown()
        goto(mid_x, mid_y)
        goto(x2, y2)


color("blue")
penup()
goto(cx, cy)
pendown()


for _ in range(20):
    forward(3)
    left(18)


penup()
goto(cx, cy + 10)
pendown()
for _ in range(20):
    forward(1.5)
    left(18)

def noha(uhel):
    penup()
    goto(cx, cy + 5)
    setheading(uhel)
    pendown()
    forward(25)
    right(20)
    forward(10)

noha(30)
noha(60)
noha(120)
noha(150)
noha(210)
noha(240)
noha(300)
noha(330)
```

```python
import random
from ColabTurtle.Turtle import *

initializeTurtle()
speed(13)

offset_x = 50
offset_y = 50

def draw_square(x, y, size, color_name):
    penup()
    goto(offset_x + x, offset_y + y)
    pendown()
    color(color_name)
    for _ in range(4):
        forward(size)
        right(90)

def draw_rectangle(x, y, width, height, color_name):
    penup()
    goto(offset_x + x, offset_y + y)
    pendown()
    color(color_name)
    for _ in range(2):
        forward(width)
        right(90)
        forward(height)
        right(90)

def draw_triangle(x, y, size, color_name):
    penup()
    goto(offset_x + x, offset_y + y)
    pendown()
    color(color_name)
    for _ in range(3):
        forward(size)
        left(120)

def draw_circle(x, y, radius, color_name):
    penup()
    goto(offset_x + x, offset_y + y - radius)
    pendown()
    color(color_name)
    for _ in range(36):
        forward(2 * 3.1415 * radius / 36)
        left(10)

colors = ["#dffbb4", "#ebb4fb", "#ffb347", "#87CEEB"]

for _ in range(50):
    shape = random.choice(["square", "rectangle", "triangle", "circle"])
    color_name = random.choice(colors)
    x = random.randint(0, 700)
    y = random.randint(0, 700)

    if shape == "square":
        size = random.randint(20, 80)
        draw_square(x, y, size, color_name)
    elif shape == "rectangle":
        width = random.randint(30, 100)
        height = random.randint(20, 80)
        draw_rectangle(x, y, width, height, color_name)
    elif shape == "triangle":
        size = random.randint(30, 100)
        draw_triangle(x, y, size, color_name)
    elif shape == "circle":
        radius = random.randint(20, 50)
        draw_circle(x, y, radius, color_name)
```
