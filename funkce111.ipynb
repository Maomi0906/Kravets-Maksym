*args a **kwargs v Pythonu umožňují funkci přijímat libovolný počet pozičních a klíčových argumentů. *args shromažďuje všechny nepojmenované (poziční) argumenty do n-tice (tuple), zatímco **kwargs sbírá pojmenované (klíčové) argumenty do slovníku. To umožňuje větší flexibilitu při definování a volání funkcí, které nepotřebují mít předem daný počet argumentů.

### *args (poziční argumenty)

Zachytí libovolný počet pozičních argumentů, které nejsou explicitně definovány v seznamu parametrů funkce.
Předá tyto argumenty jako n-tici (tuple), na kterou lze přistupovat pomocí indexů.

### **kwargs (klíčové argumenty)
Zachytí všechny nepředané klíčové argumenty do slovníku, kde klíče jsou názvy argumentů a hodnoty jsou jejich přiřazené hodnoty.

Využití - flexibilní funkce s volitelným množstvím parametrů, [dekorátory funkce](https://www.itnetwork.cz/python/oop/dekoratory-v-pythonu).

```python
def vypis_info(**kwargs):
  for klic, hodnota in kwargs.items():
    print(f"{klic}: {hodnota}")

vypis_info(jmeno="Petr", vek=30, mesto="Praha")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Устанавливаем мягкий, приглушенный стиль Seaborn
sns.set_style("whitegrid")
sns.set_palette("pastel")

filmy = [
    ("The Shawshank Redemption", 1994, 9.3),
    ("The Godfather", 1972, 9.2),
    ("The Dark Knight", 2008, 9.0),
    ("Pulp Fiction", 1994, 8.9),
    ("Forrest Gump", 1994, 8.8),
    ("Inception", 2010, 8.8),
    ("Fight Club", 1999, 8.8),
    ("Interstellar", 2014, 8.6),
    ("Parasite", 2019, 8.6),
    ("Whiplash", 2014, 8.5)
]

df = pd.DataFrame(filmy, columns=['Název', 'Rok', 'Hodnocení'])
df['Desetiletí'] = (df['Rok'] // 10) * 10

# Создаем сетку 2x2
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
fig.suptitle("Analýza filmových dat", fontsize=16, y=1.02)

# --- 1. [ВЕРХНИЙ ЛЕВЫЙ] Круговая диаграмма (из старого левого нижнего) ---
decade_counts = df['Desetiletí'].value_counts()
labels = [f'{d}-té' for d in decade_counts.index]
axs[0, 0].pie(decade_counts, labels=labels, autopct='%1.1f%%', startangle=90,
              colors=sns.color_palette("pastel"), wedgeprops={'edgecolor': 'white'})
axs[0, 0].set_title("Distribuce filmů podle desetiletí")
axs[0, 0].axis('equal')

# --- 2. [ВЕРХНИЙ ПРАВЫЙ] Линейный график (из старого правого нижнего) ---
# Возвращаем темный контрастный цвет, как в первом варианте
df_by_year = df.sort_values(by='Rok')
axs[0, 1].plot(df_by_year['Rok'], df_by_year['Hodnocení'], marker='o', linestyle='-',
               color=sns.color_palette("dark")[2], markersize=7)
axs[0, 1].set_title("Změna hodnocení s rokem vydání")
axs[0, 1].set_xlabel("Rok vydání")
axs[0, 1].set_ylabel("Hodnocení")
axs[0, 1].grid(True, alpha=0.6)

# --- 3. [НИЖНИЙ ЛЕВЫЙ] Гистограмма (из старого левого верхнего) ---
# Возвращаем исходный цвет и стиль plt.hist
axs[1, 0].hist(df['Hodnocení'], bins=5, color=sns.color_palette("pastel")[0], edgecolor="black")
axs[1, 0].set_title("Distribuce hodnocení filmů")
axs[1, 0].set_xlabel("Hodnocení")
axs[1, 0].set_ylabel("Počet filmů")
axs[1, 0].grid(axis='y', alpha=0.6)

# --- 4. [НИЖНИЙ ПРАВЫЙ] Столбчатая диаграмма (из старого правого верхнего) ---
df_sorted = df.sort_values(by='Hodnocení', ascending=True)
sns.barplot(ax=axs[1, 1], x=df_sorted['Hodnocení'], y=df_sorted['Název'],
            palette="pastel", saturation=0.8, hue=df_sorted['Název'], legend=False)
axs[1, 1].set_title("Hodnocení filmů (seřazeno)")
axs[1, 1].set_xlabel("Hodnocení")
axs[1, 1].set_ylabel("")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Устанавливаем мягкий стиль
sns.set_style("whitegrid")
sns.set_palette("pastel")

filmy = [
    ("The Shawshank Redemption", 1994, 9.3),
    ("The Godfather", 1972, 9.2),
    ("The Dark Knight", 2008, 9.0),
    ("Pulp Fiction", 1994, 8.9),
    ("Forrest Gump", 1994, 8.8),
    ("Inception", 2010, 8.8),
    ("Fight Club", 1999, 8.8),
    ("Interstellar", 2014, 8.6),
    ("Parasite", 2019, 8.6),
    ("Whiplash", 2014, 8.5)
]

df = pd.DataFrame(filmy, columns=['Název', 'Rok', 'Hodnocení'])

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
fig.suptitle("Analýza filmových dat", fontsize=16, y=1.02)

# --- 1. [ВЕРХНИЙ ЛЕВЫЙ] Круговая диаграмма ---
df['Desetiletí'] = (df['Rok'] // 10) * 10
decade_counts = df['Desetiletí'].value_counts()
labels = [f'{d}-té' for d in decade_counts.index]
axs[0, 0].pie(decade_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"), wedgeprops={'edgecolor': 'white'})
axs[0, 0].set_title("Distribuce filmů podle desetiletí")
axs[0, 0].axis('equal')

# --- 2. [ВЕРХНИЙ ПРАВЫЙ] Гистограмма рейтингов ---
sns.histplot(df['Hodnocení'], bins=5, ax=axs[0, 1], color=sns.color_palette("pastel")[2], kde=True, edgecolor="white")
axs[0, 1].set_title("Distribuce hodnocení filmů")
axs[0, 1].set_xlabel("Hodnocení")
axs[0, 1].set_ylabel("Počet filmů")

# --- 3. [НИЖНИЙ ЛЕВЫЙ] Горизонтальная диаграмма ---
df_sorted = df.sort_values(by='Hodnocení', ascending=True)
sns.barplot(ax=axs[1, 0], x=df_sorted['Hodnocení'], y=df_sorted['Název'], palette="pastel", hue=df_sorted['Název'], legend=False)
axs[1, 0].set_title("Hodnocení filmů (seřazeno)")
axs[1, 0].set_xlabel("Hodnocení")
axs[1, 0].set_ylabel("")

# --- 4. [НИЖНИЙ ПРАВЫЙ] Линейный график ---
df_by_year = df.sort_values(by='Rok')
axs[1, 1].plot(df_by_year['Rok'], df_by_year['Hodnocení'],
             marker='o', linestyle='-', linewidth=3,
             color=sns.color_palette("pastel")[3],
             markersize=10, markerfacecolor='white', markeredgewidth=2)
axs[1, 1].set_title("Změna hodnocení s rokem vydání")
axs[1, 1].set_xlabel("Rok vydání")
axs[1, 1].set_ylabel("Hodnocení")
axs[1, 1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
